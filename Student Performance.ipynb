## Preprocessing the data for further EDA:

```python
dataset['total score']=dataset['math score']+dataset['reading score']+dataset['writing score']
```

```python
from sklearn.preprocessing import LabelEncoder

# encoding the gender column
lbl_gender=LabelEncoder()
dataset['gender']=lbl_gender.fit_transform(dataset['gender'])

# encoding the race/ethnicity column
lbl_race_ethnicity=LabelEncoder()
dataset['race/ethnicity']=lbl_race_ethnicity.fit_transform(dataset['race/ethnicity'])

# encoding the race/ethnicity column
lbl_education=LabelEncoder()
dataset['parental level of education']=lbl_education.fit_transform(dataset['parental level of education'])

# encoding lunch column
lbl_lunch=LabelEncoder()
dataset['lunch']=lbl_lunch.fit_transform(dataset['lunch'])

# encoding test preparation course column
lbl_test=LabelEncoder()
dataset['test preparation course']=lbl_test.fit_transform(dataset['test preparation course'])
```
